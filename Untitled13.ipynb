```python
import pandas as pd

# Create a dictionary with Hindi sentences
data = {
    'Hindi': [
        'नमस्ते, आप कैसे हैं?',
        'मेरा नाम रोहित है।',
        'आज मौसम बहुत अच्छा है।',
        'मैं पढ़ाई कर रहा हूँ।',
        'आप कहाँ रहते हैं?',
        'मुझे हिंदी फिल्में पसंद हैं।',
        'यह एक सुंदर जगह है।',
        'क्या आप मेरी मदद कर सकते हैं?',
        'यह किताब बहुत रोचक है।',
        'मुझे क्रिकेट खेलना अच्छा लगता है।'
    ]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Save the DataFrame as a CSV file
df.to_csv('hindi_sentences_dataset.csv', index=False)

print("CSV file created successfully!")

# Download the file to your local machine
from google.colab import files
files.download('hindi_sentences_dataset.csv')
```

```python
import pandas as pd
from transformers import MarianMTModel, MarianTokenizer
from google.colab import files

# Load MarianMT model and tokenizer
def load_model_and_tokenizer(model_name='Helsinki-NLP/opus-mt-hi-en'):
    tokenizer = MarianTokenizer.from_pretrained(model_name)
    model = MarianMTModel.from_pretrained(model_name)
    return model, tokenizer

# Translate sentences in batches
def translate_batch(sentences, model, tokenizer, batch_size=32):
    translations = []
    for i in range(0, len(sentences), batch_size):
        batch = sentences[i:i+batch_size]
        encoded_batch = tokenizer(batch, return_tensors="pt", padding=True, truncation=True)
        translated_outputs = model.generate(**encoded_batch)
        decoded_batch = [tokenizer.decode(t, skip_special_tokens=True) for t in translated_outputs]
        translations.extend(decoded_batch)
    return translations

# Function to handle file upload and translation
def translate_csv_file():
    # Prompt user to upload a CSV file
    uploaded = files.upload()

    # Load the first file uploaded
    input_csv_path = next(iter(uploaded))
    df = pd.read_csv(input_csv_path)

    if 'Hindi' not in df.columns:
        raise ValueError("The uploaded CSV must contain a 'Hindi' column.")

    # Load model and tokenizer
    model, tokenizer = load_model_and_tokenizer()

    # Translate the Hindi sentences
    hindi_sentences = df['Hindi'].tolist()
    translated_texts = translate_batch(hindi_sentences, model, tokenizer)

    # Add the English translations to the DataFrame
    df['English'] = translated_texts

    # Save the translated DataFrame to a new CSV
    output_csv_path = 'translated_hindi_to_english.csv'
    df.to_csv(output_csv_path, index=False)

    print("Translation completed!")
    print("You can download the translated file using the link below.")

    # Return the file for download
    files.download(output_csv_path)
```

```python
translate_csv_file()
```
